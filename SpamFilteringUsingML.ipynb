**Step 1: Read file spam.csv**

```python
import pandas as pd

df = pd.read_csv("/content/drive/MyDrive/SpamFilteringUsingML/Data/spam.csv", encoding = 'latin1')
df.head(2)
X_words = df.iloc[:,1].values
y_label = df.iloc[:,0]
```

**Step 2: Vectorizing feature data**

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfvec=TfidfVectorizer(stop_words='english')
X=tfvec.fit_transform(X_words.astype('U')).toarray()
```

**Replace ham with 0 and spam with 1 **

```python
import numpy as np
y = np.array(y_label.replace(to_replace=['ham', 'spam'], value=[0, 1]))
```

**Step 3: We will split our dataset into 80:20 ratio where 80 for training and 20 for testing**

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0, shuffle=True)
```

**Step 4: Training with SVM model**

```python
from sklearn import svm
classifier = svm.SVC(C=10,kernel="rbf")
classifier.fit(X_train,y_train)
```

**Step 5: Predict and Accuracy**

```python
y_pred = classifier.predict(X_test)
Accuracy = 0
count = 0
for i in range(len(y_pred)):
  if y_pred[i]==y_test[i]:
    count += 1
 
Accuracy = count/len(y_pred)

print(Accuracy*100, '%')
```
